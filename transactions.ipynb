```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('transactions.csv')
df
```

What is the total sales amount across all transactions?


```python
df['t_amt'].sum()
```

Which month had the highest total transaction amount?


```python
df['t_date']=pd.to_datetime(df['t_date'],format='%d-%m-%y',errors='coerce',dayfirst=True)
df['month']=df['t_date'].dt.to_period('M')
monthly_total = df.groupby('month')['t_amt'].sum()

highest_month = monthly_total.idxmax()
highest_amount = monthly_total.max()
print(monthly_total)
highest_month, highest_amount

```

What is the average transaction amount per customer?

```python
df.groupby('cust_id')['t_amt'].mean()
```
